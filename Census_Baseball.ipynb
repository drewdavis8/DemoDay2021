```python
# US Median Household Income by State
# Request to Census.gov
census_data = c.acs5.get(("B01003_001E", "B21004_001E"), {"for": "state:*"})

# Convert to DataFrame
census_pd = pd.DataFrame(census_data)

# Column Reordering
census_pd = census_pd.rename(columns={"B01003_001E": "Population", "B21004_001E": "Median Income"})

census_pd
```
